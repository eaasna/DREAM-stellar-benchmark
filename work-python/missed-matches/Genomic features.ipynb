```python
import numpy as np
import pandas as pd
```

```python
mouse_feat = pd.read_csv("GCF_000001635.27_GRCm39_feature_table.txt", sep = "\t")
mouse_feat.drop(["assembly_unit", "seq_type", "chromosome", "assembly", "product_accession", "non-redundant_refseq", "related_accession", "symbol", "locus_tag", "product_length", "attributes", "feature_interval_length"], axis = 1, inplace = True)
mouse_feat["start"] = mouse_feat["start"].astype(int)
mouse_feat["end"] = mouse_feat["end"].astype(int)
mouse_feat.head()
```

## Not in BLAST

```python
matches = pd.read_csv("blast_masked/mouse_vs_fly_e1e-07_k12.fn.gff", sep = "\t", header = None)
matches.drop([1, 2, 5, 7], inplace = True, axis = 1)

if (len(matches[8].iloc[0].split(';')) == 4):
    matches[['qname','seq2Range','cigar', 'mutations']] = matches[8].str.split(';',expand=True)
if (len(matches[8].iloc[0].split(';')) == 5):
    matches[['qname','seq2Range', 'eValue', 'cigar', 'mutations']] = matches[8].str.split(';',expand=True)
    matches.drop(['eValue'], inplace = True, axis = 1)

matches.drop(["cigar", "mutations"], inplace = True, axis = 1)
matches.drop(["seq2Range", "qname", 8], inplace = True, axis = 1)
matches.columns = ["genomic_accession", "start", "end", "strand"]
    
matches
```

```python
matches = matches.head()
```

```python
i = 0
for index, row in matches.iterrows():
    dname = row["genomic_accession"]
    start = row["start"]
    end = row["end"]
    overlapping = mouse_feat[(mouse_feat["genomic_accession"] == row["genomic_accession"]) & (start >= mouse_feat["start"]) & (start <= mouse_feat["end"])] 
    if (len(overlapping) > 0):
        i += 1
        print("Eps-match\t" + str(dname) + "\t" + str(start) + "\t" + str(end) + "\n")
        print(overlapping)
        print()

print(str(i) + " overlap genomic features")
```
