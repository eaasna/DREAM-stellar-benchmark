```python
import numpy as np
import pandas as pd
```

```python
feat = pd.read_csv("GCF_000001635.27_GRCm39_feature_table.txt", sep = "\t")
feat.drop(["assembly_unit", "seq_type", "chromosome", "assembly", "product_accession", "non-redundant_refseq", "related_accession", "symbol", "locus_tag", "product_length", "attributes", "feature_interval_length"], axis = 1, inplace = True)
feat["start"] = feat["start"].astype(int)
feat["end"] = feat["end"].astype(int)
feat.head()
```

```python
matches = pd.read_csv("dream_masked/b1024_fpr0.005_l100_cmin0_cmax254_e1_ent0.fn.gff", sep = "\t", header = None)
matches.drop([1, 2, 5, 7], inplace = True, axis = 1)
matches[['qname','seq2Range','cigar', 'mutations']] = matches[8].str.split(';',expand=True)
matches.drop(["cigar", "mutations"], inplace = True, axis = 1)
matches.drop(["seq2Range", "qname", 8], inplace = True, axis = 1)

matches.columns = ["genomic_accession", "start", "end", "strand", "last"]
matches
```

```python
i = 0
for index, row in matches.iterrows():
    dname = row["genomic_accession"]
    start = row["start"]
    end = row["end"]
    overlapping = feat[(feat["genomic_accession"] == row["genomic_accession"]) & (start >= feat["start"]) & (start <= feat["end"])] 
    if (len(overlapping) > 0):
        i += 1
        print("Eps-match\t" + str(dname) + "\t" + str(start) + "\t" + str(end) + "\n")
        print(overlapping)
        print()

print(str(i) + " overlap genomic features")
```
