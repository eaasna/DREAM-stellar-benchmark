```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
def make_timedelta(df, col):
    df['seconds'] = df[col]
    df[col] = pd.to_timedelta(df[col], unit='s')
```

```python
size = "50Mb"

prefixes = ["blast_default", "blast", "last", "lastz", "stellar", "dist_stellar", "valik"]
table_df = pd.DataFrame([0, 0.02, 0.04, 0.06, 0.08, 0.1])
for prefix in prefixes:
    df = pd.read_csv("simulated/" + size + "/" + prefix + "_table1.tsv", sep = "\t")
    table_df[prefix + "_time"] = pd.to_timedelta(df["search time (sec)"], unit = 's')
    table_df[prefix + "_missed"] = df["missed (%)"]

table_df
```
